````python
def draw_card_grid(matrix, pid, trial, correct, condition):
    fig, ax = plt.subplots(figsize=(8, 8))
    ax.set_xlim(0, 8)
    ax.set_ylim(0, 8)
    ax.set_xticks(range(8))
    ax.set_yticks(range(8))
    ax.set_xticklabels([str(i+1) for i in range(8)])
    ax.set_yticklabels([chr(65+i) for i in range(8)])
    ax.grid(True)

    for i in range(8):
        for j in range(8):
            card = matrix[i][j]
            if card:
                color = "white"
                if "king" in card: color = "pink"
                elif "queen" in card: color = "lightblue"
                elif "jack" in card: color = "gray"
                elif "blank" in card: color = "darkgray"

                ax.add_patch(plt.Rectangle((j, 7-i), 1, 1, color=color))
                ax.text(j+0.5, 7-i+0.5, card.replace("_", "\n"),
                        ha="center", va="center", fontsize=7)

    status = "PASSED" if correct == 1 else "FAILED"
    ax.set_title(f"Participant {pid} — Trial {trial} | {status} | Condition: {condition}")
    plt.show()
````

````python
import pandas as pd
import re
import hashlib
from IPython.display import display, clear_output

# 1. LOAD AND PREP DATA
# Update 'your_file_name.csv' to match your actual file path
file_path = 'Data for LMU analyses 13 aug 2025.csv'
df = pd.read_csv(file_path, low_memory=False)

# Cleanup: Filter out junk rows and ensure Trial numbers are numeric
df = df[df['participant'].notna() & (df['participant'] != '#REF!')].copy()
df['trialN'] = pd.to_numeric(df['trialN'], errors='coerce')
df = df.dropna(subset=['trialN'])

# Identify move columns (move_0, move_1, etc.)
move_cols = [c for c in df.columns if c.startswith('move_') and 'Time' not in c]

# 2. STYLING FUNCTIONS
def get_color(name):
    """Assigns Black to blanks and pastel colors to face cards."""
    name_str = str(name).lower()
    if not name or name == '' or name_str == 'nan': return "white"

    # If the card is a blank card, return Black
    if 'blank' in name_str: return "#000000"

    # Generate a stable pastel color based on the card name
    hash_hex = hashlib.md5(name_str.encode()).hexdigest()
    r = (int(hash_hex[0:2], 16) % 100) + 160
    g = (int(hash_hex[2:4], 16) % 100) + 160
    b = (int(hash_hex[4:6], 16) % 100) + 160
    return f"rgb({r}, {g}, {b})"

def style_grid(val):
    """CSS styling for the DataFrame cells."""
    if val == '': return 'background-color: white; border: 1px solid #ddd;'
    bg = get_color(val)
    text_color = "white" if bg == "#000000" else "black"
    return f'background-color: {bg}; color: {text_color}; border: 1.5px solid #555; font-weight: bold; font-size: 9pt;'

# 3. CORE LOGIC
def show_trial(participant_id, trial_number):
    # Initialize an empty 8x8 grid
    grid = pd.DataFrame('', index=[str(i) for i in range(1, 9)], columns=list('ABCDEFGH'))

    # Get specific row
    row_data = df[(df['participant'].astype(str) == str(participant_id)) &
                  (df['trialN'] == trial_number)]

    if row_data.empty:
        print(f"No data found for Participant {participant_id}, Trial {trial_number}")
        return

    row = row_data.iloc[0]

    # Parse movement codes to fill the grid
    for col in move_cols:
        val = str(row[col])
        if val == 'nan' or val == '': continue

        # Regex to find coordinates like A1, B5, etc.
        match = re.search(r'([A-H])([1-8])', val)
        if match:
            col_coord, row_coord = match.groups()
            # Clean label (e.g., 'queen_hearts_aA1' becomes 'queen_hearts_a')
            label = val.replace(col_coord + row_coord, "").strip()
            grid.at[row_coord, col_coord] = label

    # Display Metadata
    print(f"--- TRIAL SUMMARY ---")
    print(f"PARTICIPANT: {participant_id}")
    print(f"TRIAL: {trial_number}")
    print(f"CONDITION: {row.get('condition', 'N/A')}")
    print(f"STATUS: {'✅ PASS' if str(row.get('overall_correct')) == '1.0' else '❌ FAIL'}")
    print(f"MOVEMENT CODES: {row.get('movement_codes', 'N/A')}")

    # Render Styled Grid
    display(grid.style.applymap(style_grid).set_properties(**{
        'text-align': 'center', 'width': '100px', 'height': '50px'
    }))

# 4. RUN MANUALLY
# Change these values to see different results
show_trial(participant_id='101', trial_number=0)
````

````python
# =====================================================================
# INTERACTIVE 8×8 GRID — SINGLE TRIAL VIEW + OVERLAY ALL TRIALS VIEW
# =====================================================================

import pandas as pd
import numpy as np
import ipywidgets as widgets
from IPython.display import display, HTML

# =====================================================================
# LOAD DATA
# =====================================================================
df = pd.read_csv('/content/Data for LMU analyses 13 aug 2025.csv')

# =====================================================================
# CLEAN PARTICIPANT COLUMN
# =====================================================================
df["participant"] = (
    df["participant"]
    .astype(str)
    .str.strip()
    .str.replace('.0$', '', regex=True)
)

participants = sorted(df["participant"].unique())

# =====================================================================
# COLORS FOR TRIALS
# =====================================================================
color_palette = [
    "#ff9999", "#99ff99", "#9999ff", "#ffcc99",
    "#cc99ff", "#99ffcc", "#ff99cc", "#ccff99",
    "#6699ff", "#ff6666", "#66ff66", "#6666ff",
    "#ffd700", "#00ced1", "#ff7f50", "#ba55d3"
]

def get_color(idx):
    return color_palette[idx % len(color_palette)]

# =====================================================================
# DRAW SINGLE TRIAL GRID
# =====================================================================
def draw_single_trial(participant, condition, TrialNumber):

    sub = df[
        (df["participant"] == participant) &
        (df["condition"] == condition) &
        (df["TrialNumber"] == TrialNumber)
    ]

    grid = [["" for _ in range(8)] for _ in range(8)]

    for _, row in sub.iterrows():
        r, c = int(row["row"]), int(row["col"])
        color = get_color(int(TrialNumber))
        grid[r][c] = f"""
        <div style='background:{color};
             width:40px;height:40px;
             border:1px solid black;
             display:flex;
             align-items:center;
             justify-content:center;'>
            {TrialNumber}
        </div>
        """

    html = "<table style='border-collapse:collapse;'>"
    for r in range(8):
        html += "<tr>"
        for c in range(8):
            cell = grid[r][c] if grid[r][c] else "<div style='width:40px;height:40px;border:1px solid #ccc;'></div>"
            html += f"<td>{cell}</td>"
        html += "</tr>"
    html += "</table>"

    moves = sorted(sub["movement"].astype(str).unique())
    move_dropdown = widgets.Dropdown(
        options=moves,
        description="Final Movement:",
        style={'description_width':'initial'}
    )

    display(HTML(f"<h3>Single Trial View — Trial {TrialNumber}</h3>"))
    display(HTML(html))
    display(move_dropdown)

# =====================================================================
# DRAW OVERLAY OF ALL TRIALS
# =====================================================================
def draw_overlay(participant, condition):

    sub = df[
        (df["participant"] == participant) &
        (df["condition"] == condition)
    ]

    trials = sorted(sub["TrialNumber"].unique(), key=lambda x: int(x))

    grid = [[[] for _ in range(8)] for _ in range(8)]

    # Accumulate visits from all trials
    for trial in trials:
        color = get_color(int(trial))
        trial_rows = sub[sub["TrialNumber"] == trial]

        for _, row in trial_rows.iterrows():
            r, c = int(row["row"]), int(row["col"])
            grid[r][c].append(color)

    # Build visual grid
    html = "<table style='border-collapse:collapse;'>"
    for r in range(8):
        html += "<tr>"
        for c in range(8):
            if grid[r][c]:
                # Blend colors by stacking vertically
                stacked = "".join(
                    f"<div style='background:{col};height:10px;'></div>"
                    for col in grid[r][c]
                )
                cell_html = f"""
                <div style='width:40px;height:40px;border:1px solid black;'>
                    {stacked}
                </div>
                """
            else:
                cell_html = "<div style='width:40px;height:40px;border:1px solid #ccc;'></div>"

            html += f"<td>{cell_html}</td>"
        html += "</tr>"
    html += "</table>"

    display(HTML(f"<h3>OVERLAY VIEW — ALL Trials in Condition {condition}</h3>"))
    display(HTML(html))

# =====================================================================
# MAIN UPDATE
# =====================================================================
def update_display(participant, condition, TrialNumber, mode):

    if mode == "Single Trial":
        draw_single_trial(participant, condition, TrialNumber)

    elif mode == "Overlay (All Trials)":
        draw_overlay(participant, condition)

# =====================================================================
# WIDGETS
# =====================================================================
participant_dropdown = widgets.Dropdown(
    options=participants,
    description="Participant:",
)

condition_dropdown = widgets.Dropdown(
    options=[],
    description="Condition:"
)

trial_dropdown = widgets.Dropdown(
    options=[],
    description="TrialNumber:"
)

mode_dropdown = widgets.Dropdown(
    options=["Single Trial", "Overlay (All Trials)"],
    description="View Mode:"
)

# Update condition based on participant
def on_part_change(change):
    participant = change["new"]
    conds = sorted(df[df["participant"] == participant]["condition"].unique())
    condition_dropdown.options = conds

    trials = sorted(df[df["participant"] == participant]["TrialNumber"].unique(), key=lambda x: int(x))
    trial_dropdown.options = trials

participant_dropdown.observe(on_part_change, names='value')

# Update trials when condition changes
def on_cond_change(change):
    trials = sorted(
        df[
            (df["participant"] == participant_dropdown.value) &
            (df["condition"] == change["new"])
        ]["TrialNumber"].unique(), key=lambda x: int(x)
    )
    trial_dropdown.options = trials

condition_dropdown.observe(on_cond_change, names='value')

ui = widgets.VBox([
    participant_dropdown,
    condition_dropdown,
    mode_dropdown,
    trial_dropdown
])

out = widgets.interactive_output(
    update_display,
    {
        "participant": participant_dropdown,
        "condition": condition_dropdown,
        "TrialNumber": trial_dropdown,
        "mode": mode_dropdown
    }
)

display(ui, out)

# Initialize on load
on_part_change({"new": participant_dropdown.value})
````

````python
import pandas as pd
import numpy as np
import ipywidgets as widgets
from IPython.display import display, HTML, clear_output
import ast
import re

# =====================================================================
# LOAD DATA
# =====================================================================
df = pd.read_csv('/content/Data for LMU analyses 13 aug 2025.csv')

# =====================================================================
# CARD COLORS - All blanks GREY for analysis
# =====================================================================
CARD_COLORS = {
    'queen_spades': '#FF1744',      'queen_hearts': '#E91E63',
    'queen_diamonds': '#9C27B0',    'queen_clubs': '#673AB7',
    'king_spades': '#3F51B5',       'king_hearts': '#2196F3',
    'king_diamonds': '#03A9F4',     'king_clubs': '#00BCD4',
    'jack_spades': '#009688',       'jack_hearts': '#4CAF50',
    'jack_diamonds': '#8BC34A',     'jack_clubs': '#CDDC39',
    # ALL BLANK CARDS ARE GREY - IMPORTANT FOR ANALYSIS
    'blank': '#757575', 'blank2': '#757575', 'blank3': '#757575', 'blank4': '#757575',
}

CARD_SYMBOLS = {'queen': '♛', 'king': '♚', 'jack': '♞', 'blank': '◻'}

# =====================================================================
# CLEAN DATA
# =====================================================================
df["participant"] = df["participant"].astype(str).str.strip().str.replace(r'\.0$', '', regex=True)
df["trialN"] = df["trialN"].astype(str).str.strip()

# =====================================================================
# EXTRACT CARD INFO FUNCTION
# =====================================================================
def extract_card_info(card_str):
    """Extract card type and position from strings like 'queen_spades_cA1' or 'blank_bB2'"""
    if pd.isna(card_str) or card_str == '':
        return None, None, None

    card_str = str(card_str).strip()

    # Extract card type (everything before the coordinate part)
    # Pattern: cardtype_suit_letter+coordinate OR cardtype_coordinate
    card_match = re.match(r'([a-z0-9]+(?:_[a-z]+)?)', card_str)
    card_type = card_match.group(1) if card_match else None

    # Extract coordinates (A-H)(1-8)
    coord_match = re.search(r'([A-H])([1-8])', card_str)
    if coord_match and card_type:
        col = ord(coord_match.group(1)) - ord('A')
        row = int(coord_match.group(2)) - 1
        return card_type, col, row

    return card_type, None, None

# =====================================================================
# DRAW SINGLE TRIAL GRID
# =====================================================================
def draw_single_trial(participant, condition, trialN):
    """Draw ONE 8x8 grid for ONE trial showing FINAL positions"""

    # Get the original row for this trial
    trial_row = df[
        (df["participant"] == participant) &
        (df["condition"] == condition) &
        (df["trialN"] == trialN)
    ]

    if trial_row.empty:
        return f"<p>No data found for Trial {trialN}</p>", []

    # Get pass/fail status
    overall_correct = trial_row['overall_correct'].iloc[0]
    if pd.notna(overall_correct):
        if int(overall_correct) == 1:
            status_badge = "<span style='background:green; color:white; padding:5px 15px; border-radius:5px; font-weight:bold;'>✓ PASS</span>"
        else:
            status_badge = "<span style='background:red; color:white; padding:5px 15px; border-radius:5px; font-weight:bold;'>✗ FAIL</span>"
    else:
        status_badge = "<span style='background:gray; color:white; padding:5px 15px; border-radius:5px;'>Unknown</span>"

    # Parse FINAL CARD POSITIONS (this includes blank cards!)
    final_positions = trial_row['final_card_position_codes_1'].iloc[0]
    try:
        if isinstance(final_positions, str):
            final_cards = ast.literal_eval(final_positions)
        else:
            final_cards = []
    except:
        final_cards = []

    # Also get movement codes for reference
    movement_codes = trial_row['movement_codes'].iloc[0]
    try:
        if isinstance(movement_codes, str):
            movements = ast.literal_eval(movement_codes)
        else:
            movements = []
    except:
        movements = []

    # Create 8x8 grid using FINAL POSITIONS
    grid = [[None for _ in range(8)] for _ in range(8)]

    # Place final card positions (includes blanks!)
    for card_code in final_cards:
        card_type, col, row = extract_card_info(card_code)
        if card_type and col is not None and row is not None:
            grid[row][col] = card_type

    # Count card types
    card_counts = {'queens': 0, 'kings': 0, 'jacks': 0, 'blanks': 0}
    for row in grid:
        for card in row:
            if card:
                if 'queen' in card:
                    card_counts['queens'] += 1
                elif 'king' in card:
                    card_counts['kings'] += 1
                elif 'jack' in card:
                    card_counts['jacks'] += 1
                elif 'blank' in card:
                    card_counts['blanks'] += 1

    # Build HTML
    html = f"""
    <div style='display:inline-block; margin:15px; padding:10px; border:2px solid #ddd; border-radius:8px; background:white;'>
        <h4 style='margin:5px 0; text-align:center;'>Trial {trialN} {status_badge}</h4>
        <div style='text-align:center; font-size:11px; color:#666; margin:5px 0;'>
            Movements: {len(movements)} | Cards: Q:{card_counts['queens']} K:{card_counts['kings']} J:{card_counts['jacks']} <span style='background:#757575;color:white;padding:1px 4px;border-radius:2px;'>Blank:{card_counts['blanks']}</span>
        </div>
        <table style='border-collapse:collapse; margin:5px auto;'>
            <tr><td style='border:none;'></td>
    """

    # Column headers
    for col_letter in 'ABCDEFGH':
        html += f"<td style='text-align:center; font-weight:bold; font-size:11px; padding:3px;'>{col_letter}</td>"
    html += "</tr>"

    # Grid rows
    for r in range(8):
        html += f"<tr><td style='text-align:center; font-weight:bold; font-size:11px; padding:3px; background:#f5f5f5;'>{r+1}</td>"

        for c in range(8):
            card_type = grid[r][c]

            if card_type:
                color = CARD_COLORS.get(card_type, '#999')
                rank = card_type.split('_')[0]
                symbol = CARD_SYMBOLS.get(rank, '●')

                cell_html = f"""
                <div style='background:{color}; width:45px; height:45px;
                            border:2px solid #333; border-radius:4px;
                            display:flex; align-items:center; justify-content:center;
                            color:white; font-size:24px; font-weight:bold;
                            box-shadow:2px 2px 4px rgba(0,0,0,0.3);'>
                    {symbol}
                </div>
                """
            else:
                cell_html = "<div style='width:45px; height:45px; border:1px solid #ddd; background:#fafafa;'></div>"

            html += f"<td style='padding:2px;'>{cell_html}</td>"

        html += "</tr>"

    html += "</table></div>"

    return html, movements

# =====================================================================
# DRAW ALL TRIALS
# =====================================================================
def draw_all_trials(participant, condition):
    """Draw SEPARATE grids for EACH trial"""

    # Get all trials for this participant and condition
    trials_data = df[
        (df["participant"] == participant) &
        (df["condition"] == condition)
    ]

    if trials_data.empty:
        return "<p>No data found</p>"

    trials = sorted(trials_data['trialN'].unique())

    # Count pass/fail
    passed = sum(1 for t in trials if df[(df["participant"] == participant) & (df["condition"] == condition) & (df["trialN"] == t)]['overall_correct'].iloc[0] == 1)
    failed = len(trials) - passed

    # Header
    html = f"""
    <div style='font-family:Arial; padding:15px;'>
        <h3>All Trials for Participant {participant} - Condition {condition}</h3>
        <div style='background:#fff3cd; padding:10px; border-radius:5px; margin:10px 0; border:1px solid #ffc107;'>
            <strong>📊 Data Info:</strong> This participant has <strong>{len(trials)}</strong> trial(s) in condition {condition}<br>
            <small>Trials: {', '.join(map(str, trials))}</small>
        </div>
        <div style='background:#f5f5f5; padding:10px; border-radius:5px; margin:10px 0;'>
            <strong>Summary:</strong>
            <span style='color:green; font-weight:bold;'>✓ {passed} Passed</span> |
            <span style='color:red; font-weight:bold;'>✗ {failed} Failed</span> |
            <span>Total: {len(trials)} trials</span>
        </div>
        <div style='background:#e3f2fd; padding:10px; border-radius:5px; margin:10px 0;'>
            <strong>Legend:</strong><br>
            <span style='background:#FF1744; color:white; padding:2px 8px; margin:2px; border-radius:3px;'>♛ Queens</span>
            <span style='background:#3F51B5; color:white; padding:2px 8px; margin:2px; border-radius:3px;'>♚ Kings</span>
            <span style='background:#009688; color:white; padding:2px 8px; margin:2px; border-radius:3px;'>♞ Jacks</span>
            <span style='background:#757575; color:white; padding:2px 8px; margin:2px; border-radius:3px;'>◻ Blanks (Grey)</span>
        </div>
        <div style='display:flex; flex-wrap:wrap; justify-content:flex-start;'>
    """

    # Draw each trial
    all_movements = []
    for trial in trials:
        trial_html, movements = draw_single_trial(participant, condition, trial)
        html += trial_html
        all_movements.extend(movements)

    html += "</div></div>"

    return html, all_movements

# =====================================================================
# WIDGETS
# =====================================================================
output_area = widgets.Output()

participants = sorted(df["participant"].unique())

participant_dropdown = widgets.Dropdown(
    options=participants,
    description="Participant:",
    style={'description_width': 'initial'}
)

condition_dropdown = widgets.Dropdown(
    options=[],
    description="Condition:",
    style={'description_width': 'initial'}
)

mode_dropdown = widgets.Dropdown(
    options=[("Show All Trials", "all"), ("Show Single Trial", "single")],
    value="all",
    description="View Mode:",
    style={'description_width': 'initial'}
)

trial_dropdown = widgets.Dropdown(
    options=[],
    description="Trial:",
    style={'description_width': 'initial'}
)

movement_list = widgets.Textarea(
    description="Movements:",
    disabled=True,
    layout=widgets.Layout(width='600px', height='150px'),
    style={'description_width': 'initial'}
)

def on_participant_change(change):
    participant = change["new"]
    conditions = sorted(df[df["participant"] == participant]["condition"].unique())
    condition_dropdown.options = conditions
    if conditions:
        condition_dropdown.value = conditions[0]

participant_dropdown.observe(on_participant_change, names='value')

def on_condition_change(change):
    if participant_dropdown.value and change["new"]:
        trials = sorted(
            df[
                (df["participant"] == participant_dropdown.value) &
                (df["condition"] == change["new"])
            ]["trialN"].unique()
        )
        trial_dropdown.options = trials
        if trials:
            trial_dropdown.value = trials[0]

condition_dropdown.observe(on_condition_change, names='value')

def update_display(*args):
    with output_area:
        clear_output(wait=True)

        if not participant_dropdown.value or not condition_dropdown.value:
            print("Please select participant and condition")
            return

        mode = mode_dropdown.value

        if mode == 'all':
            html, movements = draw_all_trials(
                participant_dropdown.value,
                condition_dropdown.value
            )
            display(HTML(html))

            if movements:
                movement_list.value = f"Total movements across all trials: {len(movements)}\n\nSample movements:\n" + "\n".join(movements[:20])
                if len(movements) > 20:
                    movement_list.value += f"\n... and {len(movements)-20} more"
                display(movement_list)
        else:
            trial = trial_dropdown.value
            html, movements = draw_single_trial(
                participant_dropdown.value,
                condition_dropdown.value,
                trial
            )
            display(HTML(html))

            if movements:
                movement_list.value = f"Movements for Trial {trial} ({len(movements)} total):\n\n" + "\n".join(movements)
                display(movement_list)

participant_dropdown.observe(update_display, names='value')
condition_dropdown.observe(update_display, names='value')
mode_dropdown.observe(update_display, names='value')
trial_dropdown.observe(update_display, names='value')

if participants:
    participant_dropdown.value = participants[0]

# =====================================================================
# DISPLAY UI
# =====================================================================
print("🎴 Chess Card Movement Visualizer - FIXED: Now Shows Blank Cards!")
print("=" * 70)

ui = widgets.VBox([
    widgets.HTML("<h3>Select Options:</h3>"),
    widgets.HBox([participant_dropdown, condition_dropdown]),
    widgets.HBox([mode_dropdown, trial_dropdown]),
    output_area
])

display(ui)
update_display()
````
